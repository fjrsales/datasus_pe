f) Faça uma função que identifica o capítulo do CID e a categoria a partir do código CID.

```python
## f) Identification of CID's chapter and CID's category by code
## Unified file

import chardet

def code_trat(x):
  n = len(x)
  x1 = x.replace(x[n-1], '')
  return x1

def cid_unificado():
  cid_unif = dict()
  with open('/content/drive/MyDrive/tecnicas_datasus_pe/CID-10-CAPITULOS.CSV', 'rb') as f:
      result = chardet.detect(f.read())
      
  cid_caps = pd.read_csv('/content/drive/MyDrive/tecnicas_datasus_pe/CID-10-CAPITULOS.CSV', encoding=result['encoding'], sep=';')
  cid_cat = pd.read_csv('/content/drive/MyDrive/tecnicas_datasus_pe/CID-10-CATEGORIAS.CSV', encoding=result['encoding'], sep=';')
  cid_subcat = pd.read_csv('/content/drive/MyDrive/tecnicas_datasus_pe/CID-10-SUBCATEGORIAS.CSV', encoding=result['encoding'], sep=';')

  cid_unif['CATINIC'] = cid_caps['CATINIC']
  cid_unif['CATFIM'] = cid_caps['CATFIM']
  cid_unif['DESCRICAO_CAP'] = cid_caps['DESCRICAO']
  cid_unif['CAT'] = cid_cat['CAT']
  cid_unif['DESCRICAO_CAT'] = cid_cat['DESCRICAO']
  cid_unif['SUBCAT'] = cid_subcat['SUBCAT']
  cid_unif['DESCRICAO_SUBCAT'] = cid_subcat['DESCRICAO']

  cid_unif = pd.DataFrame(cid_unif)
  cid_unif.to_csv('/content/drive/MyDrive/tecnicas_datasus_pe/CID-10_UNIFICADO.csv')


def cod_cid(code):
  cid_unificado()
  code1 = code_trat(code)
  cats = dict()
  caps = dict()
  disease = dict()

  cid_unif = pd.read_csv('/content/drive/MyDrive/tecnicas_datasus_pe/CID-10_UNIFICADO.csv')
  #display(cid_unif)

  #Chapter's, category's and subcategory's dicts from the corresponding CID
  aux = cid_unif.loc[:, ['CAT', 'DESCRICAO_CAT']]
  for cont, c in enumerate(aux['CAT']):
    cats[c] = aux['DESCRICAO_CAT'][cont]
  keys_cats = list(cats.keys())

  aux2 = cid_unif.loc[:, ['CATINIC', 'CATFIM', 'DESCRICAO_CAP']]
  
  aux3 = cid_unif.loc[:, ['SUBCAT', 'DESCRICAO_SUBCAT']]

  for cont, c in enumerate (aux2['DESCRICAO_CAP']):
    caps[c] = [aux2['CATINIC'][cont]]
    start = keys_cats.index(aux2['CATINIC'][cont])
    end = keys_cats.index(aux2['CATFIM'][cont])
    for j, value in enumerate(cats):
      if (j > start) and (j <= end):
        caps[c].append(value)
  
  #print(cats)
  #print(caps)
  
  #for k,v in caps.items():
   # print(f'{k}: {v}')

  #Identification of disease by CID's code
  disease['CODIGO'] = code
  for k,v in caps.items():
    if code1 in v:
      disease['CAPÍTULO'] = k
      break
  for k, v in cats.items():
    if code1 == k:
      disease['CATEGORIA'] = v
      break
  for cont, v in enumerate (aux3['SUBCAT']):
    if v == code:
      disease['SUBCATEGORIA'] = aux3['DESCRICAO_SUBCAT'][cont]
      break
  return disease

  
print(cod_cid('E149'))  #Example
```

g) Determine as 10 maiores causas de morte por ano em ordem decrescente, trazendo o número de mortes por causa e a porcentagem em relação ao total de mortos no ano/período

```python
## The 10 greatest death causes per year in descending order, bringing the number 
## of deaths per cause and the percentage related to the total of deaths per year/period.

deaths_per_year = dict()
deaths_per_year_percentage = dict()
df3 = df2.loc[:, ['CAUSABAS', 'ANO_O']]
#display(df3)
years = list(range(2010,2020))
total_per_year = list()

for y in years:
  df3_filtered = df3[df3['ANO_O'] == y]
  deaths_per_year[y]= dict(df3_filtered['CAUSABAS'].value_counts(sort=True))
  total_per_year.append(sum(deaths_per_year[y].values()))

for k1 in deaths_per_year.keys():
  count = 0
  aux = dict()
  for k2,v2 in deaths_per_year[k1].items():
    aux[k2] = v2
    count += 1
    if count == 10:
      break
  deaths_per_year[k1] = aux

print('The 10 greatest death causes per year in descending order:')
for k,v in deaths_per_year.items():
  print(f'{k} = {v}')
```

```python
## The 10 greatest death causes per year in descending order, bringing the number 
## of deaths per cause and the percentage related to the total of deaths per year/period.
## Graphics

import matplotlib.pyplot as plt

for k,v in deaths_per_year.items():
  illness_legend = list()

  for k2,v2 in deaths_per_year[k].items():
    ill = cod_cid(k2)
    illness_legend.append(f"{k2}: {ill['SUBCATEGORIA']}")

  plt.figure(figsize=(16,8))
  plt.title(f'The 10 highest death causes in {k}')
  plt.xlabel('Death causes in CID codes')
  plt.ylabel(f'Number of deaths')
  plt.bar(deaths_per_year[k].keys(), deaths_per_year[k].values())
  
  x = 5000
  for i in illness_legend:
    plt.text(3, x, i, fontsize='small')
    x = x - 200
  plt.savefig(f'/content/drive/MyDrive/tecnicas_datasus_pe/Images/The 10 highest death causes in {k}.png')

```

```python
## Percentage related to the total of deaths per year

for k1 in deaths_per_year.keys():
  cont2 = 0
  deaths_per_year_percentage[k1] = dict(deaths_per_year[k1])
  for k2,v2 in deaths_per_year[k1].items():
    deaths_per_year_percentage[k1][k2] = round((v2/(total_per_year[cont2]/100)),2)
  cont2 += 1

print('Percentage related to the total of deaths per year')
for k,v in deaths_per_year_percentage.items():
  print(f'{k} = {v}')
```

h) Verifique se houve mudança no ranking, ano a ano.

```python
sorted_codes = []
for k,v in deaths_per_year_percentage.items():
  sorted_codes.append(sorted(deaths_per_year_percentage[k]))
  print(f'{k} = {sorted_codes[-1]}')
```

```python
def highlight_rows(x):
    if x.Cause_of_Death_Following_Year != x.Cause_of_Death_Current_Year:
        return['background-color: purple']*3
    else:
        return["background-color: ''"]*3



years = list(deaths_per_year_percentage.keys())
for i,y in zip(range(len(sorted_codes)),years):
  table = {}
  table["Ranking"] = []
  table["Cause_of_Death_Current_Year"] = []
  table["Cause_of_Death_Following_Year"] = []
  if i < (len(sorted_codes)-1) and list(sorted_codes[i]) == list(sorted_codes[i+1]):
    print(f"From {y} to {y+1} there were no changes in the ranking of top 10 causes of death.")
    for j,k in zip(sorted_codes[i], sorted_codes[i+1]):
      table["Ranking"].append(str(list(sorted_codes[i]).index(j)+1)+"º")
      table["Cause_of_Death_Current_Year"].append(CID_dict[j])
      table["Cause_of_Death_Following_Year"].append(CID_dict[k])
    tdf = pd.DataFrame(data=table)
    display(tdf.style.apply(highlight_rows, axis = 1))

  elif i < (len(sorted_codes)-1):
    print(f"From {y} to {y+1} there were changes in the ranking of top 10 causes of death:")
    for j,k in zip(sorted_codes[i], sorted_codes[i+1]):
      table["Ranking"].append(str(list(sorted_codes[i]).index(j)+1)+"º")
      table["Cause_of_Death_Current_Year"].append(CID_dict[j])
      table["Cause_of_Death_Following_Year"].append(CID_dict[k])
      if j != k:
        try:
          print(f"{j} is in #{list(sorted_codes[i]).index(j)+1} in {y} and it is in #{list(sorted_codes[i+1]).index(j)+1} in {y+1}")
        except:
          print(f"{j} is in #{list(sorted_codes[i]).index(j)+1} in {y} and it is not of the top 10 causes of death in {y+1}")
    tdf = pd.DataFrame(data=table)
    display(tdf.style.apply(highlight_rows, axis = 1))

  print("\n")
```
