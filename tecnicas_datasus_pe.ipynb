```python
## For each column, it determines the percentage of null content
df_list = list(df.keys())
for col in df_list[:]:
  total_null=df[col].isnull().sum()
  print('Column {} has total null {}, percentage of {}% of content is null.'.format(col,total_null,round(total_null*100/len(df[col]),2)))
```
